##### 1. Please use decision tree regression and neural network regression to see if they improve the performances of throughput  as compared with the previous homework assignments.

```python
import pandas as pd 
df = pd.read_csv("throughput.csv")
df.head(2)
```

```python
df.info()
```

```python
throughput = df.dropna()
throughput.info()
```

```python
throughput['X6']
```

```python
from sklearn.preprocessing import MinMaxScaler
ss = MinMaxScaler()
X_scaled = ss.fit_transform(throughput.iloc[:, 6:])
pd.DataFrame(X_scaled, columns = throughput.columns[ 6 :])
```

```python
cat_df = pd.concat([throughput['X1'], throughput['X2'], throughput['X3'], throughput['X4'], throughput['X5']], axis=1)
cat_df = pd.get_dummies(cat_df).astype('int64')

cat_df.head(2)
```

```python
data = pd.concat([pd.DataFrame(X_scaled, columns = throughput.columns[ 6 :]), 
                  pd.DataFrame(cat_df.values, columns = cat_df.columns[0:]),
                  pd.DataFrame(throughput['Y'].values, columns = [throughput.columns[ 0 ]])], axis = 1) 

data = data.reset_index(drop=True)
data.head(2)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], test_size=0.2, random_state=42)
```

(a) decision tree

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg.fit(X_train, y_train)
```

```python
tree_reg.predict(X_test)
```

```python
from statsmodels.tools.eval_measures import rmse 
from sklearn.metrics import r2_score

y_test_pred = tree_reg.predict(X_test)
print('Test RMSE:' , round(rmse(y_test, y_test_pred), 4))
print('Test R^2:' , round(r2_score(y_test, y_test_pred), 4))

y_train_pred = tree_reg.predict(X_train)
print('Train RMSE:' , round(rmse(y_train, y_train_pred), 4))
print('Train R^2:' , round(r2_score(y_train, y_train_pred), 4))
```

(b) neural network

```python
import keras
from keras import models
from keras import layers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64, activation='relu',
                           input_shape=(X_train.shape[1],)))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model
```

```python
model = build_model()
model.fit(X_train, y_train,
          epochs=80, batch_size=16, verbose=0)
test_mse_score, test_mae_score = model.evaluate(X_test, y_test)
train_mse_score, train_mae_score = model.evaluate(X_train, y_train)
```

##### 2. Please use decision tree classification and neural network classification to see if they improve the performances of ad clicking as compared with the previous homework assignments.

```python
import pandas as pd
click = pd.read_excel("HW4_click_16k.xlsx")
click.head(2)
```

```python
click.info()
```

```python
X_cat = click.drop(columns = ["click"])
X_cat = X_cat.applymap(str)
X_cat.info()
```

```python
X_cat_ohe = pd.get_dummies(X_cat)
X_cat_ohe.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_cat_ohe, click.iloc[:, 1], test_size=0.30)
```

(a) decision tree

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg.fit(X_train2, y_train2)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_test2_pred = tree_reg.predict(X_test2)
print('Test RMSE:' , round(mean_squared_error(y_test2, y_test2_pred), 4))
print('Test R^2:' , round(r2_score(y_test2, y_test2_pred), 4))

y_train2_pred = tree_reg.predict(X_train2)
print('Train RMSE:' , round(mean_squared_error(y_train2, y_train2_pred), 4))
print('Train R^2:' , round(r2_score(y_train2, y_train2_pred), 4))
```

(b) neural network

```python
import keras
from keras import models
from keras import layers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64, activation='relu',
                           input_shape=(X_train2.shape[1],)))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model

model = build_model()
model.fit(X_train2, y_train2,
          epochs=80, batch_size=16, verbose=0)
test_mse_score, test_mae_score = model.evaluate(X_test2, y_test2)
train_mse_score, train_mae_score = model.evaluate(X_train2, y_train2)
```
